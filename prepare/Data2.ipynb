```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/Admin/Desktop/DA1/data/DataStroke_OK.csv")
data.head()
```

```python
data["gender"] = data["gender"].map({0:"Female", 1:"Male"})
```

```python
data["ever_married"] = data["ever_married"].map({1:"Yes",0:"No"})
```

```python
data["work_type"] = data["work_type"].map({0:"Private" ,1:"Self-employed"  , 2:"children" , 3:"Govt_job",4: "Never_worked"})
```

```python
data["smoking_status"]= data["smoking_status"].map({0:"never smoked" ,1: "Unknown" , 2: "formerly smoked"  , 3:"smokes"})
```

```python
data["residence_type"] = data["residence_type"].map({0:"Urban", 1:"Rural"})
```
