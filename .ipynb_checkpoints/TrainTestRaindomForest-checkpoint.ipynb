```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
#-----------
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
data_ptRDFR = pd.read_csv('data_stroke_ok.csv')
data_ptRDFR.head()
```

```python
#L·∫•y gi√° tr·ªã bi·∫øn ƒë·ªôc l·∫≠p: X
X = data_ptRDFR.loc[:,['gender', 'age', 'hypertension', 'heart_disease', 
                      'ever_married','work_type', 'residence_type', 'avg_glucose_level', 'bmi', 'smoking_status']]

print('K√≠ch th∆∞·ªõc bi·∫øn ƒë·ªôc l·∫≠p X:',X.shape)
X
```

```python
#L·∫•y gi√° tr·ªã bi·∫øn ph·ª• thu·ªôc: y
y=data_ptRDFR.loc[:,'stroke']

print('K√≠ch th∆∞·ªõc bi·∫øn ph·ª• thu·ªôc y:',y.shape)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, #Bi·∫øn ƒë·ªôc l·∫≠p X t√°ch th√†nh (X_train; X_Test)
                                                    y, #Bi·∫øn ph·ª• thu·ªôc y t√°ch th√†nh (y_train; y_test)
                                                    train_size=0.8, #T·ª∑ l·ªá t√°ch: train = 0.8 (80%)
                                                    shuffle=True)   #L·∫•y c√°c m·∫´u ng·∫´u nghi√™n 

print("T·ªïng s·ªë h·ªì s∆° ban ƒë·∫ßu:",len(X))
print('1.K√≠ch th∆∞·ªõc t·∫≠p Training:', X_train.shape)
print('2.K√≠ch th∆∞·ªõc t·∫≠p Test:', X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Kh·ªüi t·∫°o m√¥ h√¨nh
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)

# Hu·∫•n luy·ªán
rf_model.fit(X_train, y_train)

# D·ª± ƒëo√°n
y_pred = rf_model.predict(X_test)

# ƒê√°nh gi√° k·∫øt qu·∫£
print("ƒê·ªô ch√≠nh x√°c:", accuracy_score(y_test, y_pred))
print("\nB√°o c√°o ph√¢n lo·∫°i:\n", classification_report(y_test, y_pred))
print("\nMa tr·∫≠n nh·∫ßm l·∫´n:\n", confusion_matrix(y_test, y_pred))
```

```python
# Tinh ch·ªânh m√¥ h√¨nh
rf_tuned = RandomForestClassifier(
    n_estimators=400,        # tƒÉng s·ªë c√¢y
    max_depth=15,            # gi·ªõi h·∫°n ƒë·ªô s√¢u
    min_samples_split=5,     # y√™u c·∫ßu √≠t nh·∫•t 5 m·∫´u ƒë·ªÉ chia node
    min_samples_leaf=3,      # √≠t nh·∫•t 3 m·∫´u trong 1 l√°
    max_features='sqrt',     # ch·ªçn ng·∫´u nhi√™n sqrt(n_features) ƒë·∫∑c tr∆∞ng
    class_weight='balanced', # c√¢n b·∫±ng d·ªØ li·ªáu
    random_state=42,
    n_jobs=-1                # d√πng to√†n b·ªô CPU
)

# Hu·∫•n luy·ªán
rf_tuned.fit(X_train, y_train)

# D·ª± ƒëo√°n
y_pred_tuned = rf_tuned.predict(X_test)

# ƒê√°nh gi√°
print("üéØ ƒê·ªô ch√≠nh x√°c:", accuracy_score(y_test, y_pred_tuned))
print("\nüìä B√°o c√°o ph√¢n lo·∫°i:\n", classification_report(y_test, y_pred_tuned))
print("\nüìâ Ma tr·∫≠n nh·∫ßm l·∫´n:\n", confusion_matrix(y_test, y_pred_tuned))
```
